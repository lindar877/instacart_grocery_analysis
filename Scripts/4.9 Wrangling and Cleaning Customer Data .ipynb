```python
# Import libraries 

import pandas as pd
import numpy as np
import os
```

```python
# importing by using os.path.join() function 

path = r'/Users/lindazhang/Instacart Basket Analysis'
df_cust = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'customers.csv'), index_col = False)
df_ords_prods_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```

## 01. Consistency Checks for customer data set

```python
df_cust.shape
```

```python
df_cust.head()
```

```python
# change column names

df_cust.rename(columns = {'user_id' : 'customer_id'}, inplace = True)
df_cust.rename(columns = {'Surnam' : 'last_name'}, inplace = True)
```

```python
# identify data types of each column 

df_cust.dtypes
```

```python
# change the data type of date_joined column

df_cust['date_joined'] = df_cust['date_joined'].astype('datetime64[ns]')
```

```python
df_cust['date_joined'].dtype
```

```python
df_cust.head()
```

```python
# Checking the data 

df_cust.describe()
```

```python
# Checking for mixed data type columns

for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].applymap(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

```python
# Since the First Name column has mixed data types, we can change it all to string 

df_cust['First Name'] = df_cust ['First Name'].astype('str')
```

```python
# Checking for missing values...looks like there are none. 

df_cust.isnull().sum()
```

```python
# checking for duplicates...looks like there are none. 

df_cust_dups = df_cust[df_cust.duplicated()]
df_cust_dups
```

## 02. Checking the ords_prods_merged data set

```python
df_ords_prods_merged.shape
```

```python
# looking at the df_ords_prods_merged dataframe

df_ords_prods_merged.head(20)
```

```python
# Dropping a few columns from ords_prods_merged


df_ords_prods_merged = df_ords_prods_merged.drop(columns = ['order_added_to_cart', 'reordered', 'aisle_id'])
```

## 03. Merging customer and ords_prods_merged data sets

```python
# merging with ords_prods_merged

df_ords_prods_all = df_ords_prods_merged.merge(df_cust, on = ['customer_id'], indicator = True)
```
