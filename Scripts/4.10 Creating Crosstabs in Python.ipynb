Crosstabs are a common tool for conducting data checks in Python. Think of them as Python’s version of Excel’s pivot tables.

```python
# Import libraries 

import pandas as pd
import numpy as np
import os
```

```python
# importing by using os.path.join() function 

path = r'/Users/lindazhang/Instacart Basket Analysis'
ords_prods_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```

```python
# this is the code you'd use to create a crosstab

crosstab = pd.crosstab(ords_prods_merged['days_since_prior_order'], ords_prods_merged['order_number'], dropna = False)
```

```python
crosstab
```

You can see above that where order number is 1, all the days_since_prior order are 0 which makes sense. 

```python
# this is how you can copy and paste your crosstab so you can view it in Excel

crosstab.to_clipboard()
```
